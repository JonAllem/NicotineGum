```python
import pandas as pd
topic_details = pd.read_excel('../data/retry2_topics_nic_gum_text_data_2022.xlsx', sheet_name='Sheet1').set_index('Topic #')
topic_details.head()
```

```python
data = pd.read_parquet('../data/modelled/modelled_nic_gums_tweets_2022.parquet')
data.head()
```

```python
exclude = pd.read_excel('../data/Combined Validation Try 3.xlsx',sheet_name='SheetCorrect')
exclude.head()
```

```python
data.index
```

```python
topic_details.Category.nunique()
```

```python
exclude_categories = ['Smoking Cessation','Others']
# data.drop(exclude.id.apply(str).to_list(), axis = 0, inplace=True)
data.drop_duplicates(subset=['text'], inplace=True)
topic_details['Category'] = topic_details['Category'].apply(lambda x: x.strip())
print(data.shape)
data['category'] = data['original_topic'].apply(lambda x: topic_details.loc[x, 'Category'])
print(data.category.nunique())
data_sample_only_25 = data[~data.category.isin(exclude_categories)].groupby('category', group_keys=False).apply(lambda x: x.sample(n=25))
data_sample_only_25.category.value_counts()
```

```python
data_sample_only_25.to_excel('../data/nic_gum_25_tweetsptopic_no_smoking_cessation.xlsx', index=True)
```

```python
n_samples = 1005
print(data.shape)
data.drop(exclude.id.apply(str).to_list(), axis = 0, inplace=True)
print(data.shape)
data['category'] = data['original_topic'].apply(lambda x: topic_details.loc[x, 'Category'])
data_sample = data[data.category!='Others'].groupby('category', group_keys=False).apply(lambda x: x.sample(n_samples * len(x) // len(data[data.category!='Others'])))
data_sample.category.value_counts(), data_sample.category.value_counts().sum()
```

```python
data_sample.category.value_counts().sum()
```

```python
data_sample = data_sample.reset_index()[['category','id', 'text']].sort_values('category', ascending=False)
data_sample.head()
```

```python
data_sample.to_excel('../data/nic_gum_1000_sample_try3.xlsx', index=False)
```

```python
data_sample_25 = data_sample.sample(frac=0.25).sort_values('category')
data_sample_25.to_excel('../data/nic_gum_250_sample_try3.xlsx', index=False)
```
