```python
import pandas as pd
data = pd.read_parquet('../data/modelled/modelled_nic_gums_tweets_2022.parquet')
data.info()
```

```python
topic_details = pd.read_excel('../data/retry2_topics_nic_gum_text_data_2022.xlsx', sheet_name='Sheet1').set_index('Topic #')
topic_details.head()
```

```python
topic_details.groupby('Category').agg({'Topic Size (Tweet Count)':'sum'})
```

```python
double_coded = pd.read_excel('../data/Double Coded Validation Final.xlsx', sheet_name='Sheet1')
double_coded
```

```python
double_coded = double_coded.dropna(subset=['id'])
double_coded.info()
```

```python
double_coded.category.value_counts()
```

```python
topic_details['Category'] = topic_details['Category'].apply(lambda x: x.strip())
print(data.shape)
data['category'] = data['original_topic'].apply(lambda x: topic_details.loc[x, 'Category'])
data.category.value_counts()
```

```python
data.reset_index(inplace=True)
data['id'] = data['id'].astype('float64')
double_coded['id'] = double_coded['id'].astype('float64')
data_not_in_double_coded = data[~data.id.isin(double_coded.id)]
data_not_in_double_coded.info()
```

```python
n_samples = 1004 - double_coded.shape[0]
(data_not_in_double_coded[data_not_in_double_coded.category!='Others'].groupby('category', group_keys=False).apply(lambda x: x.sample(n_samples * len(x) // len(data_not_in_double_coded[data_not_in_double_coded.category!='Others']))).category.value_counts() + double_coded.category.value_counts()).sum()
```

```python
data_remaining_sample = data_not_in_double_coded[data_not_in_double_coded.category!='Others'].groupby('category', group_keys=False).apply(lambda x: x.sample(n_samples * len(x) // len(data_not_in_double_coded[data_not_in_double_coded.category!='Others'])))
data_remaining_sample
```

```python
assert data_remaining_sample.id.isin(double_coded.id).sum() == 0
```

```python
data_remaining_sample.to_excel('../data/nic_gum_validation_remaining_sample_580.xlsx', index=False)
```
