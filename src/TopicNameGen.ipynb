```python
import pandas as pd
topic_df = pd.read_excel('../data/topics_nic_gum_text_data_2022.xlsx')
topic_df.head()
```

```python
data = pd.read_parquet('../data/modelled/modelled_nic_gums_tweets_2022.parquet')
data.info()
```

```python
sample_text_list = ['I have performed Topic Modeling on tweets about Nicotine Gum. Use the following sample of tweets and the top words to give 1 name for the topic. \
                    Find patterns in the tweets (like what is the main theme they are talking about). \
                    You can explain the reason for the name in detail. \
                    Mention if the theme does not seem to be primarily about the US and \
                      mention names of the countries outside of the US in the topic name if this is the case. \
                    Make sure the theme you capture, and the topic name emphasize the relation to Nicotine Gum of that theme. \
                    If you deem correct, give alternate topic names as well. Write the topic name suitable for mentioning in a research paper.\
                    \nTweets -\n']
for i,x in enumerate(data[data.original_topic==1].text.sample(30)): sample_text_list.append(f'Tweet {i+1}\n{x}')
# print('\n'.join(sample_text_list))
pro = '\n'.join(sample_text_list)
response = openai.Completion.create(
  model="text-davinci-003",
  prompt=f"\"\"\"\n{pro}\n\"\"\"\n",
  temperature=0,
  max_tokens=64,
  top_p=1.0,
  frequency_penalty=0.0,
  presence_penalty=0.0,
  stop=["\"\"\""]
)
print(response['choices'][0]['text'])
```
